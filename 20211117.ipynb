```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 100.0     # sine frequency, Hz
phase = np.pi/2  # phase, radian
#설정값 세팅
```

```python
# generate time
nSamp = sr * dur
t = np.arange(1, nSamp+1)/sr    #t에 1에서 nSamp까지의 어레이를 sampling rate로 나눈 값을 넣음(시간값)
```

```python
# generate phase
theta = t * 2*np.pi * freq + phase  #각도 = 시간 * 360 * sin freq + 90 (90도 부터)
```

```python
# generate signal by cosine-phasor
s = amp * np.sin(theta) #사인 함수의 range 설정
```

```python
plt.plot(t, s, '.') #x축에 t, y축에 s 
plt.xlabel('time (s)')
plt.ylabel('real')#라벨 설정
```

```python
ipd.Audio(s, rate=sr) #s를 설정된 sampling rate로 음성 출력
```

```python
def phasor_sine(amp, sr, dur, freq, phase):

  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)

  return s, t, theta
```

```python
amp = 1; sr = 10000; dur = 1; freq = 440; phase = 0; #값 설정
s, t, theta = phasor_sine(amp, sr, dur, freq, phase) #phasor sine 함수에서 리턴된 값을 s, t, theta에 할당
ipd.Audio(s, rate=sr) #출력
```
